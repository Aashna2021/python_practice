```python
n=3
for i in range(0,n):
    print('i='+str(i), end=" ")
    for j in range(0,5):
        if(j == n//2 or j == n//2-i or j == n//2+i):
            print('*',end="|")
        else:
            print(' ',end="|")
    print('')
```
